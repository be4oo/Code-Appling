Task is to build a game to play with the computer a guessing number game

Consists of 3 levels
- In the first level, you have `3` trials to guess the number that the computer randomly has in a range from `1:10`
- In the second level, you have `7` trials to guess the number that the computer randomly has in a range from `1:100`
- In the third level, you have `15` trials to guess the number that the computer randomly has in a range from `1:1000`



```python
# Flags for the game

easy = medium = hard = False

guessedRight = False

# Make the user choose the level

print("Welcome to the Guessing Number Game")
print("Please choose the level you want to play")
print("1. Easy")
print("2. Medium")
print("3. Hard")
print("Please enter the number of the level?")

level = int(input())

if level == 1:
    easy = True
    print("Thank you, you choose number 1 which is the easy level")
elif level == 2:
    medium = True
    print("Thank you, you choose number 2 which is the medium level")
elif level == 3:
    hard = True
    print("Thank you, you choose number 3 which is the hard level")
else:
    print("Invalid level")
    
# Make the computer choose a random number

import random

if easy:
    number = random.randint(1, 10)
    print("Now, I have put my number in my mind, try to put your first guess")
elif medium:
    number = random.randint(1, 100)
    print("Now, I have put my number in my mind, try to put your first guess")
elif hard:
    number = random.randint(1, 1000)
    print("Now, I have put my number in my mind, try to put your first guess")

# print(number)

# Get the user's guessing number

if easy:
    for i in range (1,4):
        guess = int(input())
        print("Your guess is " + str(guess))
        if guess == number:
            print("Congratulation, you have guessed the number right")
            guessedRight = True
            break
        elif guess > number:
            print("Sorry, you have guessed bigger number")
        else:
            print("Sorry, you have guessed smaller number")
elif medium:
    for i in range (1,7):
        guess = int(input())
        print("Your guess is " + str(guess))
        if guess == number:
            print("Congratulation, you have guessed the number right")
            guessedRight = True
            break
        elif guess > number:
            print("Sorry, you have guessed bigger number")
        else:
            print("Sorry, you have guessed smaller number")
elif hard:
    for i in range (1,15):
        guess = int(input())
        print("Your guess is " + str(guess))
        if guess == number:
            print("Congratulation, you have guessed the number right")
            guessedRight = True
            break
        elif guess < number:
            print("Sorry, you have guessed bigger number")
        else:
            print("Sorry, you have guessed smaller number")

# If the user fails to guess it right in his trials tell him your number

if not guessedRight:
    print("Sorry, you have failed to guess the number")
    print("My number was " + str(number))
```

```python
import tkinter as tk
from tkinter import messagebox
import random

# Create the main window
root = tk.Tk()
root.title("Guessing Number Game")

# Variables
level = tk.StringVar()
guess = tk.StringVar()
number = 0
guesses = 0
max_guesses = 0

# Functions
def set_level():
    global number, guesses, max_guesses
    guesses = 0
    if level.get() == "1":
        number = random.randint(1, 10)
        max_guesses = 3
    elif level.get() == "2":
        number = random.randint(1, 100)
        max_guesses = 6
    elif level.get() == "3":
        number = random.randint(1, 1000)
        max_guesses = 15
    else:
        messagebox.showerror("Error", "Invalid level")

def check_guess():
    global guesses
    guesses += 1
    try:
        user_guess = int(guess.get())
        if user_guess == number:
            messagebox.showinfo("Congratulations", "You have guessed the number right")
        elif user_guess > number:
            messagebox.showinfo("Hint", "You have guessed a bigger number")
        elif user_guess < number:
            messagebox.showinfo("Hint", "You have guessed a smaller number")
        if guesses == max_guesses:
            messagebox.showinfo("Game Over", f"Sorry, you have failed to guess the number. My number was {number}")
    except ValueError:
        messagebox.showerror("Error", "Please enter a valid number")

# Widgets
tk.Label(root, text="Please choose the level you want to play").pack()
tk.Radiobutton(root, text="Easy", variable=level, value="1").pack()
tk.Radiobutton(root, text="Medium", variable=level, value="2").pack()
tk.Radiobutton(root, text="Hard", variable=level, value="3").pack()
tk.Button(root, text="Set Level", command=set_level).pack()

tk.Label(root, text="Enter your guess").pack()
tk.Entry(root, textvariable=guess).pack()
tk.Button(root, text="Check Guess", command=check_guess).pack()

# Start the event loop
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import random

# Function to start the game based on selected level
def start_game(level):
    global number, attempts
    if level == "Easy":
        number = random.randint(1, 10)
        attempts = 3
    elif level == "Medium":
        number = random.randint(1, 100)
        attempts = 6
    elif level == "Hard":
        number = random.randint(1, 1000)
        attempts = 14
    
    level_label.config(text=f"Level: {level}")
    attempts_label.config(text=f"Attempts left: {attempts}")
    guess_button.config(state=tk.NORMAL)
    guess_entry.config(state=tk.NORMAL)
    result_label.config(text="")

# Function to handle user's guess
def make_guess():
    global attempts
    try:
        guess = int(guess_entry.get())
        attempts -= 1
        attempts_label.config(text=f"Attempts left: {attempts}")
        
        if guess == number:
            result_label.config(text="Congratulations, you have guessed the number right!")
            end_game()
        elif guess > number:
            result_label.config(text="Sorry, you guessed a bigger number.")
        else:
            result_label.config(text="Sorry, you guessed a smaller number.")
        
        if attempts == 0 and guess != number:
            result_label.config(text=f"Sorry, you have failed to guess the number. My number was {number}")
            end_game()
    except ValueError:
        messagebox.showerror("Invalid input", "Please enter a valid number")

# Function to end the game
def end_game():
    guess_button.config(state=tk.DISABLED)
    guess_entry.config(state=tk.DISABLED)

# Setup the main window
window = tk.Tk()
window.title("Guessing Number Game")

# Add widgets
tk.Label(window, text="Welcome to the Guessing Number Game").pack(pady=10)
tk.Label(window, text="Please choose the level you want to play:").pack(pady=5)

level_frame = tk.Frame(window)
level_frame.pack(pady=5)
tk.Button(level_frame, text="Easy", command=lambda: start_game("Easy")).grid(row=0, column=0, padx=5)
tk.Button(level_frame, text="Medium", command=lambda: start_game("Medium")).grid(row=0, column=1, padx=5)
tk.Button(level_frame, text="Hard", command=lambda: start_game("Hard")).grid(row=0, column=2, padx=5)

level_label = tk.Label(window, text="Level: None")
level_label.pack(pady=5)

attempts_label = tk.Label(window, text="Attempts left: N/A")
attempts_label.pack(pady=5)

tk.Label(window, text="Enter your guess:").pack(pady=5)
guess_entry = tk.Entry(window, state=tk.DISABLED)
guess_entry.pack(pady=5)

guess_button = tk.Button(window, text="Submit Guess", state=tk.DISABLED, command=make_guess)
guess_button.pack(pady=5)

result_label = tk.Label(window, text="")
result_label.pack(pady=5)

# Start the main event loop
window.mainloop()
```
