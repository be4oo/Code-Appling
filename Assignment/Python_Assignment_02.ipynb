# **Exercise 2:**

Write a program to accept three numbers from the user and print:
Min. Number and Max. Number

```python
x = []
for i in range (0,3):
    x.append(int(input("Please enter a number: ")))


min = 100000
max = 0

for i in x:
    if i > max:
        max = i

print("Max. Number is " + str(max))

for i in x:
    if i < min:
        min = i

print("Min. Number is " + str(min))
```

# **Exercise 5:**

Write a program to chek if a word is palindrome (read from right as read from left, as "eye", "radar" and "madam"). 
> Print `True` or `False`

```python
word = input("Enter a word: ")

is_palindrome = True

for i in range(len(word)):
    if word[i] != word[len(word)-i-1]:
        is_palindrome = False
        break

#is_palindrome = word == word[::-1]

print(is_palindrome)
```

# **Exercise 7:**

Write a program to calculate how many sentances and words in the following paragraph

```python
paragraph = '''Hello Friends. This is a light exercise. Enjoy!'''

# calculate No. of Sentences & No. of words

sentences = paragraph.count('.') + paragraph.count('!') + paragraph.count('?')  # count sentences

words = paragraph.count(' ') + 1  # count words 

print("No. of Sentences: " + str(sentences))

print("No. of Words: " + str(words))
```
