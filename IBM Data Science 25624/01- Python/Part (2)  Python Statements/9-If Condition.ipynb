### Multiple Branches


```python
num1 = float(input('Enter a number: '))
num2 = float(input('Enter another number: '))

if num1 > num2:
    print('The first number is greater than the second number')
elif num1 < num2:
    print('The first number is smaller than the second number')
else:
    print('The first number is equal to the second number')
```

### Simple Calculator

```python
num1 = float(input('Enter a number: '))
num2 = float(input('Enter another number: '))
operator= input('Enter the required operation:\n"+" for adding,"-" for subtracting, "*" for multiplication, "/" for dividing: ' )

if operator =='+':
    result = num1 + num2
elif operator =='-':
    result = num1 - num2
elif operator =='*':
    result = num1 * num2
elif operator =='/':
    result = num1 / num2
else:
    result = 'False operation'
print(result)
```

### Nested if

```python
num = int(input('Enter an integer number: '))
if num >= 10:
   if num <=100:
        print("Between 10 and 100")
   else:
        print('More than 100')
else:
    print('Less than 10')

print('Done')
        
```

```python
num = int(input('Enter an integer number: '))
if (num >= 10 and num <=100):
    print("Between 10 and 100")
elif num > 100:
    print('More than 100')
else:
    print('Less than 10')

print('Done')
```
