### List Comprehension

```python
l=[]
for letter in 'Hello':
    l.append(letter)
l    
```

```python
list('Hello')
```

```python
[letter for letter in 'Hello']
```

```python
[letter.upper() for letter in 'Hello']
```

```python
[num for num in range(11)]
```

```python
list(range(11))
```

```python
[num**2 for num in range(1,11)]
```

```python
# list of even numbers from 1 to 50
l = []
for n in range(1,51):
    if n % 2 == 0:
        l.append(n)

l
```

```python
[n for n in range(1,51) if n %2 == 0]
```

```python
[n for n in range(0,51,2)]
```

```python
lst=[]
for num in range(1,11):
    lst.append(num**2)
lst    
```

```python
[num**2 for num in range(11) if num % 2 == 0]
```

```python
# Convert Celsius to Fahrenheit
celsius = [20,25,27,28,30,35]

fahrenheit = [((9/5)* t + 32) for t in celsius ]

fahrenheit
```
