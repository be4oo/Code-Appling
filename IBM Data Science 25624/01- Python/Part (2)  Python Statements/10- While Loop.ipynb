#### break
> Breaks out of the current closest enclosing loop.


```python
num =1    
while num <=10:
    if num == 6:
        break
    print(num)
    num +=1   
     
print('Done') 
```

#### continue
> Goes to the top of the closest enclosing loop

```python
num =1    
while num <=10:
    if num == 6:
        num+=1
        continue
    print(num)
    num +=1    
print('Done') 
```

```python
num =1    
while num <=10:
    if num == 6:
        num +=1
        continue   
    else:
        print(num)
        num +=1    
print('Done') 
```

```python
num = 0    
while num < 10:
    num +=1
    if num == 6:
        continue   
    print(num)   
print('Done') 
```

#### pass
>  Does nothing at all

```python
num =1    
while num <=10:
    if num == 6:
        pass
    print(num)
    num +=1    
print('Done') 
```
