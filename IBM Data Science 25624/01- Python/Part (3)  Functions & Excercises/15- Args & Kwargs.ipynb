# Part (3): Functions and Modules

### \*Args & **Kwargs:

* (\*args): You can pass undefined length of a sequence arguments using *param_name

* (\*\*kwargs): You can pass undefined length of key-value pairs using **param_name

```python
def average(n1,n2,n3):
    return (n1+n2+n3)/3
```

```python
average(5,10,15)
```

```python
def mean(*nums):
    return sum(nums)/len(nums)
```

```python
mean(1,2,3,4,5)
```

```python
mean(3,5,20,30,40,50)
```

```python
mean(*(range(100)))
```

```python
# Sum of numbers from 1 to 1000
sum(range(1,1001))
```

```python
# Maximum of any collection of numbers
max([2,4,5,9,12])
```

```python
def maximum(*nums):
    return max(nums)
```

```python
maximum(2,4,9,1312,45754,232)
```

```python
def data(**user):
    return user['hope']
```

```python
data(name='ahmed', age= 31, address= 'ABC', hope= 'football')
```

```python
def subjects(**subjects):
    math_science = subjects['math'] + subjects['science']
    ar_en = subjects['arabic'] + subjects['english']
    if math_science > ar_en:
        return 'scientific'
    else:
        return 'arts'

subjects(arabic= 10, english= 20, math= 30, science= 40, history= 50)
```
