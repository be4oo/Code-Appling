### User-defined Functions

```python
# Function Definition
def absolute(num):   # num : Parameter
    if num >= 0:
        return num
    else:
        return -num    
```

```python
# Function Calling
absolute(4)   # 4: Argument
```

```python
absolute(-5)
```

```python
print("""
    Function circle_area : calculates the area of circle
    Input: radius: the radius of circle
    Return: the area of circle 
      """)
```

```python
def circle_area(radius):
    """
    Function circle_area : calculates the area of circle
    Input: radius: the radius of circle
    Return: the area of circle
    """
    return 3.14 * (radius**2)
```

```python
circle_area(10)
```

```python
print(circle_area.__doc__)
```

```python
def circle_area(radius, PI = 3.14):
    """
    Function circle_area : calculates the area of circle
    Input: radius: the radius of circle
    Return: the area of circle
    """
    return PI * (radius**2)
```

### Sample Calculator:

```python
def calculate(num1,num2,operator):
    """
    Simple Calculator
    Input: num1: First Number
           num2: Second Number
           operator: one of the following mathematical operations: '+', '-', '*', '/'
    Return: the result of thwe mathematical operation
    """
    if operator =='+':
        result = num1 + num2
    elif operator =='-':
        result = num1 - num2
    elif operator =='*':
        result = num1 * num2
    elif operator =='/':
        result = num1 / num2
    else:
        result = 'False operation'

    return(result)
```

```python
calculate(5,10,'*')
```

```python
calculate(23,12,'/')
```
