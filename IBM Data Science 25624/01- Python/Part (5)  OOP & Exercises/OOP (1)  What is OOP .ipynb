## Intro

> Python is OOP Language (**O**bject **O**riented **P**rogramming)

        * Everything is object

```python
a = 10
b = 3.25
c = "hi"
d = [1,2,3,5]
print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
a = 0         # different instance of <class 'int'>
b = 15.55     # different instance of <class 'float'>
c = "welcome" # different instance of <class 'str'>
d = [1,2]     # different instance of <class 'list'>
print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
s1 ='hi'
s2 = 'bye'

# same method for any instance of <class 'str'>
print(s1.upper()) 
print(s2.upper())   
```

 *  **So, we have predefined data types (classes)**
  * *Every class can have many instances*
  * *Any class has predefined methods (functions)*
