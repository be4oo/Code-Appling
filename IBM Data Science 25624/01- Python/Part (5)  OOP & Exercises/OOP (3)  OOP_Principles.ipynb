# Abstraction & Encapsulation

```python
class Employee():
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def annual_raise(self, kpi_1, kpi_2, kpi_3):
        new_salary = self.salary + kpi_1 * 500 + kpi_2 * 1.5 + kpi_3 / 10
        if new_salary > 20000:
            tax = new_salary * 0.2
        else:
            tax = new_salary * 0.15

            # Many details may be here
        medical_insurance = new_salary * 0.10
        social_insurance = new_salary * 0.15
        self.salary = new_salary - (medical_insurance + social_insurance + tax) 
        return self.salary  
```

```python
emp1 = Employee('Anas', 10000)
```

```python
emp1.name  # Attribute
```

```python
emp1.salary   # Attribute
```

```python
# Abstraction
emp1.annual_raise(25, 120, 150)   # Method
```

```python
emp1.salary
```

# Inheritence

*  **When we create classes we can inherit methods and attributes from other already existing classes.** 

*  **This allows us to create a variety of different sub-classes or child classes based on what is known as a parent class or super class.**



```python
class Manager(Employee):
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
```

```python
m1 = Manager("Ali", 40000)
```

```python
m1.annual_raise(50, 120, 150)  # inherited from Employee
```

```python
class Accountant(Employee):
    
    def __init__(self, name, salary):
        super().__init__(name, salary)
        

    # Overriding
    def annual_raise(self, kpi_1, kpi_2, kpi_3):
        new_salary = self.salary + kpi_1 * 500 + kpi_2 * 1.5 + kpi_3 / 10
        if new_salary > 20000:
            tax = new_salary * 0.1
        else:
            tax = new_salary * 0.05

            # Many details may be here
        medical_insurance = new_salary * 0.13
        social_insurance = new_salary * 0.09
        self.salary = new_salary - (medical_insurance + social_insurance + tax) 
        return self.salary 
```

```python
emp1 = Employee('Anas', 10000)
emp1.annual_raise(25, 120, 150)
```

```python
acc = Accountant('Anas', 10000)
acc.annual_raise(25, 120, 150)
```

 If we create a method or attribute inside of our child class with the same name as one in the parent it will override the parent class (and may have a different functionality).

# Polymorphism

polymorphism.png

```python
class Tennis_Player():
    def shoot(self):
        pass
```

```python
class Football_Player():
    def shoot(self):
        pass
```

```python
class Basketball_Player():
    def shoot(self):
        pass
```

```python
class Baseball_Player():
    def shoot(self):
        pass
```

```python
p1 = Tennis_Player()
p2 = Football_Player()
p3 = Basketball_Player()
p4 = Baseball_Player()
```

```python
# Same Name , Different Behavior
p1.shoot()
p2.shoot()
p3.shoot()
p4.shoot()
```
