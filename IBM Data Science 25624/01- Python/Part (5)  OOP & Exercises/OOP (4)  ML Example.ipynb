## Data Splitting

````python
features = data['data']
target = data['target'].reshape(-1,1)
````

````python
features.shape
````

````python
target.shape
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target)
````

## Modelling

````python
# Linear Regression
from sklearn.linear_model import LinearRegression    # import Class (blueprint)
reg_model1 = LinearRegression()                      # Make an instance (copy)
reg_model2 = LinearRegression()
reg_model3 = LinearRegression()
````

````python
reg_model1.fit(X_train, y_train)                           # Fitting Method
````

````python
# reg_model1.predict(X_test)                                 # Prediction Method
````

````python
reg_model1.score(X_test, y_test)                           # Scoring Method
````

````python
reg_model1.coef_            # Attribute
````

````python
# Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
reg2 = DecisionTreeRegressor()

reg2.fit(X_train, y_train)                       # Fitting Method  (Different Behavior)
reg2.predict(X_test)                             # Prediction Method  (Different Behavior)
reg2.score(X_test, y_test)                       # Scoring Method  (Different Behavior)
````

# OOP Principles
1. *Abstraction:*
> Hide unnecessary details from user
2. *Encapsulation:*
> Combine Attributes and Methods
3. *Polymorphism:*
> Many Functions with the same name, but different beahviour
4. *Inheritence:*
> Create new classes from existing classes
