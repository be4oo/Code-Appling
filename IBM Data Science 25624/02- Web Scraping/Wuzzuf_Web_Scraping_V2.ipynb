```python
# Importing Libraries
# Install necessary libraries for web scraping
# !pip install lxml
# !pip install requests
# !pip install bs4

# Importing necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Fetching Webpage Content
# Fetching the webpage content
res = requests.get('https://wuzzuf.net/search/jobs/?q=data%20analysis&a=hpb')
soup = BeautifulSoup(res.content, 'html.parser')

# Display the prettified version of the HTML content
print(soup.prettify())
```

```python
# Cell 3: Finding Specific Elements
# Finding specific elements using BeautifulSoup
soup.find('h2', {'class':'css-m604qf'})
soup.find('h2', class_='css-m604qf').text
soup.find_all('h2', class_='css-m604qf')
```

```python
# Extracting Job Titles and Links
# Extracting job titles
jobs = soup.find_all('h2', class_='css-m604qf')
jobs_ =[job.text for job in jobs]

# Extracting job links
jobs_link_ = [job.a['href'] for job in jobs]
```

```python
# Extracting Job Posting Time and Occupation
# Extracting job posting time
time = soup.find_all('div', class_= 'css-1lh32fc')
time_ = [job.find_all('a')[0].text.strip() for job in time]

# Extracting job occupation
occupation = soup.find_all('span', class_= 'css-o1vzmt eoyjyou0')
occupation_ = [job.text.strip() for job in occupation]
```

```python
# Extracting Company Names and Locations
# Extracting company names
company = soup.find_all('div', class_= 'css-d7j1kk')
company_ = [c.a.text.replace(' -', '') for c in company]

# Extracting locations
location = soup.find_all('div', class_= 'css-d7j1kk')
location_ = [l.span.text.strip() for l in location]
```

```python
# Checking the length
print(len(jobs_))
print(len(jobs_link_))
print(len(time_))
print(len(occupation_))
print(len(company_))
print(len(location_))
```

```python
# Creating a DataFrame
# Creating a DataFrame from the extracted information
df = pd.DataFrame({'Jobs': jobs_, 'Company': company_, 'Location': location_, 'Link': jobs_link_, 'Time': time_, 'Occupation': occupation_})
df
```

```python
# Saving the DataFrame to an Excel File
# Saving the DataFrame to an Excel file
df.to_excel('wuzzuf.xlsx', index=False)
```

```python
# Scraping Multiple Pages
# Function to scrape Wuzzuf for a given query and number of pages
def wuzzuf(query, pages=3):
    jobs_lst = []
    companies_lst = []
    locations_lst = []
    links_lst = []
    times_lst = []
    occupations_lst = []

    for i in range(pages):
        url = f'https://wuzzuf.net/search/jobs/?q={query}&start={i}'
        res = requests.get(url)
        soup = BeautifulSoup(res.content, 'lxml')
        jobs = soup.find_all('h2', class_='css-m604qf')
        jobs_ =[job.text.strip() for job in jobs]
        company = soup.find_all('div', class_= 'css-d7j1kk')
        company_ = [c.a.text.replace(' -', '') for c in company]
        location = soup.find_all('div', class_= 'css-d7j1kk')
        location_ = [l.span.text.strip() for l in location]
        jobs_link_ = [job.a['href'] for job in jobs]
        time = soup.find_all('div', class_= 'css-1lh32fc')
        time_ = [t.find_all('a')[0].text.strip() for t in time]
        occ_ =[]
        for t in time:
                try:
                        occ_.append(t.find_all('a')[1].text.strip() )
                except:
                        occ_.append('NA')

        jobs_lst.extend(jobs_)
        companies_lst.extend(company_)
        locations_lst.extend(location_)
        links_lst.extend(jobs_link_)
        times_lst.extend(time_)
        occupations_lst.extend(occ_)

    df = pd.DataFrame({'Jobs': jobs_lst, 'Company': companies_lst, 'Location': locations_lst, 'Link': links_lst, 'Time': times_lst, 'Occupation': occupations_lst})
    return df
```

```python
# Writing the Scraping Function to a Python File
# Writing the scraping function to a Python file
%%writefile wuzzuf.py

import pandas as pd
import requests
from bs4 import BeautifulSoup

def get_jobs(query, pages=3):
    jobs_lst = []
    companies_lst = []
    locations_lst = []
    links_lst = []
    times_lst = []
    occupations_lst = []

    for i in range(pages):
        url = f'https://wuzzuf.net/search/jobs/?q={query}&start={i}'
        res = requests.get(url)
        soup = BeautifulSoup(res.content, 'lxml')
        jobs = soup.find_all('h2', class_='css-m604qf')
        jobs_ =[job.text.strip() for job in jobs]
        company = soup.find_all('div', class_= 'css-d7j1kk')
        company_ = [c.a.text.replace(' -', '') for c in company]
        location = soup.find_all('div', class_= 'css-d7j1kk')
        location_ = [l.span.text.strip() for l in location]
        jobs_link_ = [job.a['href'] for job in jobs]
        time = soup.find_all('div', class_= 'css-1lh32fc')
        time_ = [t.find_all('a')[0].text.strip() for t in time]
        occ_ =[]
        for t in time:
                try:
                        occ_.append(t.find_all('a')[1].text.strip() )
                except:
                        occ_.append('NA')

        jobs_lst.extend(jobs_)
        companies_lst.extend(company_)
        locations_lst.extend(location_)
        links_lst.extend(jobs_link_)
        times_lst.extend(time_)
        occupations_lst.extend(occ_)

    df = pd.DataFrame({'Jobs': jobs_lst, 'Company': companies_lst, 'Location': locations_lst, 'Link': links_lst, 'Time': times_lst, 'Occupation': occupations_lst})
    df.to_excel(f'wuzzuf_jobs_{query}.xlsx', index=False)
    return df.head()

if __name__ == '__main__':
    print('Wuzzuf Jobs')
    query = input('Enter Job Title: ')
    pages = int(input('Enter No. of Pages: '))
    print(f'Jobs for {query} are being scraped:')
    df = get_jobs(query, pages)
    print(f'wuzzuf_jobs_{query} file has been created')
    print('First 5 jobs:')
    print(df)
```
