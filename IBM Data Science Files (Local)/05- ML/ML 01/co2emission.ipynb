
## Understanding the Data

### `FuelConsumption.csv`:
We have downloaded a fuel consumption dataset, **`FuelConsumption.csv`**, which contains model-specific fuel consumption ratings and estimated carbon dioxide emissions for new light-duty vehicles for retail sale in Canada. [Dataset source](http://open.canada.ca/data/en/dataset/98f1a129-f628-4ce4-b24d-6f16bf24dd64)

- **MODELYEAR** e.g. 2014
- **MAKE** e.g. Acura
- **MODEL** e.g. ILX
- **VEHICLE CLASS** e.g. SUV
- **ENGINE SIZE** e.g. 4.7
- **CYLINDERS** e.g 6
- **TRANSMISSION** e.g. A6
- **FUEL CONSUMPTION in CITY(L/100 km)** e.g. 9.9
- **FUEL CONSUMPTION in HWY (L/100 km)** e.g. 8.9
- **FUEL CONSUMPTION COMB (L/100 km)** e.g. 9.2
- **CO2 EMISSIONS (g/km)** e.g. 182   --> low --> 0

# One Feature

```python
X = df[['ENGINESIZE']]   
y = df[['CO2EMISSIONS']]
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

Linear Equation

Co2 Emission = 126.2 + 39.3 * Engine Size

## OLS From Scratch

```python
import numpy as np

# Add a column of ones to X for the intercept term
X_b = np.hstack([np.ones((X.shape[0], 1)), X])   # We Use the whole X not only X_train
```

```python
X.shape, X_b.shape
```

```python
X_b
```

```python
# Calculate the coefficients using the normal equation
weights = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
weights
```

## Evaluation

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
print('Mean Absolute Error: ', mean_absolute_error(y_test, y_pred))
print('Accuracy: ', r2_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop= 'first', sparse_output=False).set_output(transform='pandas')
X_train_enc = ohe.fit_transform(X_train[cat])
X_test_enc = ohe.transform(X_test[cat])
```

```python
# Scaling for numerical
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler().set_output(transform='pandas')
X_train_num = scaler.fit_transform(X_train[imp_features])
X_test_num = scaler.transform(X_test[imp_features])
```

```python
X_train_num
```

```python
X_train_num.min(), X_train_num.max()
```

```python
X_train_all = pd.concat([X_train_num, X_train_enc], axis=1)
X_train_all
```

```python
X_test_all = pd.concat([X_test_num, X_test_enc], axis=1)
X_test_all
```

```python
reg2 = LinearRegression()
reg2.fit(X_train_all, y_train)
```

```python
y_pred = reg2.predict(X_test_all)

print('Mean Absolute Error: ', mean_absolute_error(y_test, y_pred))
print('Accuracy: ', r2_score(y_test, y_pred))
```
