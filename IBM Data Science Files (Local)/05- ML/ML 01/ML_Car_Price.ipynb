- **car_ID**: Unique identifier for each car.
- **symboling**: Risk factor rating assigned to the car (-3 to +3).
- **CarName**: Name of the car.
- **fueltype**: Type of fuel used by the car (e.g., gas, diesel).
- **aspiration**: Type of aspiration used in the engine (e.g., std, turbo).
- **doornumber**: Number of doors on the car (e.g., two, four).
- **carbody**: Body style of the car (e.g., sedan, hatchback).
- **drivewheel**: Type of drive wheel (e.g., fwd, rwd, 4wd).
- **enginelocation**: Location of the engine (e.g., front, rear).
- **wheelbase**: Distance between the front and rear wheels.
- **enginesize**: Size of the engine in cubic inches.
- **fuelsystem**: Type of fuel system (e.g., mpfi, 2bbl).
- **boreratio**: Ratio of the bore (diameter of the cylinder) to the stroke.
- **stroke**: Length of the piston stroke.
- **compressionratio**: Ratio of the volume of the combustion chamber from its largest capacity to its smallest capacity.
- **horsepower**: Power output of the engine.
- **peakrpm**: Maximum revolutions per minute (RPM) of the engine.
- **citympg**: Fuel efficiency in city driving (miles per gallon).
- **highwaympg**: Fuel efficiency in highway driving (miles per gallon).
- **price**: Price of the car.

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
print('Mean Absolute Error: ', mean_absolute_error(y_test, y_pred))
print('Accuracy: ', r2_score(y_test, y_pred))
```

```python
def process_new_data(X_test):
    # Ensure X_test is a DataFrame
    if not isinstance(X_test, pd.DataFrame):
        X_test = X_test.to_frame().T
    
    # Process numerical features
    X_test_num = scaler.transform(X_test[imp_features])
    
    # Process categorical features
    X_test_enc = ohe.transform(X_test[cat])
    
    # Concatenate numerical and categorical features
    X_test_all = pd.concat([X_test_num, X_test_enc], axis=1)
    
    return X_test_all
```
