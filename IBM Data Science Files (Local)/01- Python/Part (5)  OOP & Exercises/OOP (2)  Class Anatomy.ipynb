## Attributes

```python
class Customer: # class is a blueprint for an object
    
    def __init__(self, name='new_customer', age=25):  # Constructor (called automatically when an object is created)
        self.name = name # Instance variable
        self.age = age  # self is the instance of the class
```

```python
c1 = Customer('Anas', 5) 
```

```python
c1.name
```

```python
c1.age
```

> `name` , `age` are attributes of our class (properties >>> Variables)

```python
c1 = Customer("Ali", 30)
```

```python
c1.name
```

```python
c1.age
```

## Methods

```python
class Customer: # class is a blueprint for an object
    
    def __init__(self, name, age): 
        self.name = name 
        self.age = age  

    def place_order(self): # Instance method
        print(f"{self.name} have placed an order")

    def cancel_order(self): # Instance method
        print(f"{self.name} have cancelled an order")
   
```

> `place_order` and `cancel_order` are methods of our class (Behaviors >>> functions)

```python
c1 = Customer("Mohamed", 30)
```

```python
c1.place_order()
```

```python
c1.cancel_order()
```

```python
type(c1)
```

```python
c1.age
```

```python
isinstance(c1, Customer)
```
