## Set Operations

```python
set1 = {1,2,3}
set2 = {3,4,5}
```

```python
set1.union(set2)
```

```python
set1.intersection(set2)       
```

```python
set2.difference(set1)          
```

```python
set1.issubset(set2)  
```

```python
set1.issuperset(set2)               
```
