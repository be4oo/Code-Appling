### Scope

```python
v=5

def No_change_v(num):
    v = num + 10    # Local Change
    return v
```

```python
No_change_v(50)
```

```python
v
```

```python
def change_v(num):
    global v
    v = num + 10    # Global Change
    return v
```

```python
change_v(50)
```

```python
v
```

```python
x = 5

def foo():

    x = 10
    print("local x:", x)


foo()
print("global x:", x)
```
