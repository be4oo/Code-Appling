```python
def swap_case(s):
    new_s=''
    for c in s:
        if c.islower():
            new_s += c.upper()
        elif c.isupper():
            new_s += c.lower()     
        else:
            new_s += c       
    return new_s

swap_case('Zero-Grad.com presents "Pythonist"')
    
```

        Write a function to check prime number


# A4

```python
def is_prime(num):
    prime = True
    for i in range(2,num):
        if num%i == 0:
            prime = False
            break
    return prime

is_prime(13)            
```

```python
def is_valid(password):
    alpha_small_flag = False
    alpha_cap_flag = False
    nums_flag = False

    symbols = '@#$'
    symbols_flag = False
    
    valid_flag = True
    
    if len(password) < 6 : 
        print("Your password must be at least 6 characters")
    elif len(password) > 12 :
        print("Your password must be at Max. 12 characters")  
    else:
        for c in password:
            if c.islower() and not alpha_small_flag:
                alpha_small_flag = True
            elif c.isupper() and not alpha_cap_flag:
                alpha_cap_flag = True
            elif c.isnumeric() and not nums_flag:
                nums_flag = True
            elif c in symbols and not symbols_flag:
                symbols_flag = True
            elif valid_flag:
                continue
            else:
                print(f"{c} is not valid character") 
                valid_flag = False               

        if not alpha_small_flag:
            print("Your Password must include Small Letter")
        if not alpha_cap_flag:    
            print("Your Password must include Capital Letter") 
        if not nums_flag:
            print("Your Password must include a number")
        if not symbols_flag:    
            print("Your Password must include a symbol(@,#,$)")
        if alpha_small_flag and alpha_cap_flag and nums_flag and symbols_flag and valid_flag:
            print("Acceptable Password")     

is_valid('sA1ZaAAA##ss')
```

# Q6
        

        Write a function to check review is positive, negative or neutral by counting number of positive words and negative words

* positive_lst = ['great', 'wonderful', 'happy', 'amazing', 'attractive', 'fantastic', 'fun']
* negative_lst = ['bad', 'terrible', 'horrible', 'silly']

# A6

```python
def check_review(review):
    positive_lst = ['great', 'wonderful', 'happy', 'amazing', 'attractive', 'fantastic', 'fun']
    negative_lst = ['bad', 'terrible', 'horrible', 'silly']

    pos_count = 0
    neg_count = 0

    for word in review.split():
        if word in positive_lst:
            pos_count = pos_count + 1
        elif word in negative_lst:
            neg_count += 1

    if pos_count > neg_count:
        print("Positive Review")
    elif pos_count < neg_count:
        print("Negative Review")
    else:
        print("Neutral Review")            
```

```python
review1 = '''Professional programmers aren't going to learn anything new from this book, but it was a fun read regardless.
            What struck me as great about this book, is it reads like an ELI5 about algorithms and data structures.
            If nothing else, it's a great way for one to learn how to teach algorithms to others who aren't in the industry. 
            This is a must read for anyone that's learning to code.'''

check_review(review1)            
```
