### zip
>* __Syntax__ : zip (*iterators)
* __Return Value__ :Returns a single iterator object, having mapped values from all the containers.

```python
number_list = [1, 2, 3]
list(enumerate(number_list))
```

```python
for i, number in enumerate(number_list):
    print(i, '>>>>', number)
```

```python
number_list = [1, 2, 3]
index_list = [0,1,2]
list(zip(index_list, number_list))
```

```python
number_list = [1, 2, 3]
str_list = ['one', 'two', 'three']
l = []
for i in range(len(number_list)):
    l.append((number_list[i], str_list[i]))
l
```

```python
number_list = [1, 2, 3]
str_list = ['one', 'two', 'three']
[(number_list[i], str_list[i]) for i in range(len(number_list))]
```

```python
number_list = [1, 2, 3]
str_list = ['one', 'two', 'three']
list(zip(number_list, str_list))
```

```python
for num, st in zip(number_list, str_list):
    print(num,'>>>>', st)
```

```python
combined_list= list(zip(number_list, str_list))
print(combined_list)
```

```python
dict(zip(number_list, str_list))
```

```python
number_list = [1, 2, 3]
str_list = ['one', 'two', 'three']
d = dict()
for i in range(len(number_list)):
    d[str_list[i]] = number_list[i]

d
```

```python
# Capitalize every word in a sentence
msg = 'welcome to python course'
' '.join([word.capitalize() for word in msg.split()])
```

```python
def capitalize_sent(sentence):
    return ' '.join([word.capitalize() for word in sentence.split()])


capitalize_sent('welcome to python course')
```

### Filter

filter ( func , iterator)

```python
# Program to filter out only the even items from a list
my_list = [1, 5, 4, 6, 8, 11, 3, 12]

new_list = list(filter(lambda x: (x%2 == 0) , my_list))

print(new_list)
```

```python
# Program to filter out only the even items from a list
my_list = [1, 5, 4, 6, 8, 11, 3, 12]

def is_even(num):
    return num % 2 == 0

new_list = list(filter(is_even , my_list))

print(new_list)
```

### Map

map ( func , iterator)

```python
# Program to double each item in a list using map()

my_list = [1, 5, 4, 6, 8, 11, 3, 12]

new_list = list(map(lambda x: x * 2 , my_list))

print(new_list)
```

```python
# Program to double each item in a list using map()

my_list = [1, 5, 4, 6, 8, 11, 3, 12]

def double(num):
    return num * 2

new_list = list(map(double , my_list))

print(new_list)
```

### Reduce

```python
# importing functools for reduce()
from functools import reduce
 
# initializing list
nums = [1, 5 , 10 , 20 ]
 
# using reduce to compute sum of list
sum = reduce(lambda x,y: x+y , nums)
print(sum)
 
# using reduce to compute maximum element from list
max= reduce(lambda x, y: x if x>y else y, nums)
print(max)

```
