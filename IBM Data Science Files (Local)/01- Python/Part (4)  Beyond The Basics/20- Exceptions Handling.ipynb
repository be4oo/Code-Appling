```python
def division_fn(num1, num2):
    try: 
        return num1/num2
    except (ZeroDivisionError, TypeError) as err:
        return(f'error: {err}')

print(division_fn(1,'0'))
print(division_fn(1,0))
print(division_fn(1,4))
```

```python
def sqrt(x):
    assert x > 0, 'must be positive number'
    try:
        return x ** 0.5
    except TypeError:
        print('x must be an int or float')

sqrt(-0.5)        
```

```python
def sqrt(x):
    try:
        # assert x > 0
        return x ** 0.5
    except TypeError:
        print('x must be an int or float')
    except AssertionError:
        print('No')

sqrt(-0.5)        
```
