```python
list(enumerate(['C++', 'Java', 'Python']))
```

```python
for i in ['C++', 'Java', 'Python']:
    print(i)
```

```python
list(enumerate(['C++', 'Java', 'Python']))
```

```python
for i in enumerate(['C++', 'Java', 'Python']):
    print(i)
```

```python
for index, element in enumerate(['C++', 'Java', 'Python']):
    print(index)
```

```python
for index, element in enumerate(['C++', 'Java', 'Python']):
    print(element)
```

```python
for index, element in enumerate(['C++', 'Java', 'Python']):
    print(f'Language: {element} at index: {index}')
```

```python
for index, element in enumerate(['C++', 'Java', 'Python'], start =1):
    print(f'Language: {element} at index: {index}')
```

```python
for index, element in enumerate(('C++', 'Java', 'Python'), start =100):
    print(f'Language: {element} at index: {index}')
```

```python
d= dict()
d['C++'] = 101
d['Java'] = 102
```

```python
d = dict()
for index, element in enumerate(('C++', 'Java', 'Python'), start =100):
    d[element] = index 

d    
```
