### range
>range(start,stop,step)
* default values are start = 0 and step = 1 
* loop until value is stop - 1

```python
# print Numbers from 1 to 10
for num in range (1,11):
    print(num)
```

```python
for num in range (1,5):
    if num == 3:
        print('Hi')
    print(num)
```

```python
for num in range (1,5):
    if num == 3:
        print('Hi')
    else:    
        print(num)
```

```python
for num in range (1,11):
    if num == 6:
        continue
    print(num)
```

```python
for num in range (1,11):
    if num == 6:
        break
    print(num)
```

### Calculator again ! but *modified*





```python
while True:
    num1 = float(input('Enter a number: '))
    num2 = float(input('Enter another number: '))
    operator= input('Enter the required operation:"+" for adding,"-" for subtracting, "*" for multiplication, "/" for dividing: ' )

    if operator =='+':
        print('{} {} {} = {}'.format(num1,operator,num2, num1+num2))
    elif operator =='-':
        print('{} {} {} = {}'.format(num1,operator,num2, num1-num2))
    elif operator =='*':
        print('{} {} {} = {}'.format(num1,operator,num2, num1*num2))
    elif operator =='/':
        print('{} {} {} = {}'.format(num1,operator,num2, num1/num2))
    else:
        print('False operation')
        
    repeat= input(' Enter "y" for another operation or another key to break: ')
    if repeat =='y':
        continue
    else:
        break

```

### for loop with different data structures:

```python
# with strings
for _ in 'Hello':
    print(_)
```

```python
# with lists
for i in [2,4,6,8]:
    print(i)
```

```python
# with tuples
tup = ((2,4),[6,8],(10,12))
for i in tup:
    print(i)
```

```python
# with dictionaries:
dic= {'Name': 'Anas', 'Age': 4, 'Level': 1}
```

```python
for i in dic:
    print(i)
```

```python
for i in dic.keys():
    print(i)
```

```python
for i in dic.values():
    print(i)
```

```python
for i in dic.items():
    print(i)
```

```python
for (k,v) in dic.items():
    print(k)
```

```python
for k,v in dic.items():
    print(v)
```

```python
for k,v in dic.items():
    print('{} >> {}'.format(k,v))
```
